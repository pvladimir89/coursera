```python
from math import exp,sin
def func1(x):
    return sin(x/5)*exp(x/10)+5*exp(-x/2)
```

```python
import numpy as np
from scipy import linalg
arr_coord=np.arange(1.,15.1,0.1)
arr_funct=np.array([func1(coord) for coord in arr_coord])
```
